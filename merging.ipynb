## Let's first try just 1 indicator df with area 

### Step 1) Since the place names may be represented differently in different datasets (e.g. "Kota A" vs "A", "AB" vs "A B"), create a matching dictionary!

```python
# Make all province, kabupaten, kota names in uppercase for both columns in the 2 datasets to prevent differences and failing to merge
area['Kabupaten/kota'] = area['Kabupaten/kota'].str.upper()
melted_pop['Province/Regency/City'] = melted_pop['Province/Regency/City'].str.upper()
```

```python
import difflib

# Create a list of town names from each dataset
towns_melted = melted_pop['Province/Regency/City'].tolist()
towns_area = area['Kabupaten/kota'].tolist()

# Find close matches between town names
matches_melted_to_area = {}
matches_area_to_melted = {}

# Find matches from melted_pop to area
for town_melted in towns_melted:
    match = difflib.get_close_matches(town_melted, towns_area, n=1)
    if match:
        matches_melted_to_area[town_melted] = match[0]

# Find matches from area to melted_pop
for town_area in towns_area:
    match = difflib.get_close_matches(town_area, towns_melted, n=1)
    if match:
        matches_area_to_melted[town_area] = match[0]

# Print the matches
print("Matches from melted_pop to area:")
for town_melted, match_area in matches_melted_to_area.items():
    print(f"{town_melted} -> {match_area}")
```

```python
len(matches_melted_to_area)
```

#### Issue 1) Duplicate matches

```python
# Create a list to store duplicate matches
duplicate_matches = []

# Check for duplicates
for key, value in matches_melted_to_area.items():
    if list(matches_melted_to_area.values()).count(value) > 1:
        duplicate_matches.append((key, value))

# Print duplicate matches
print("Duplicate Matches:")
for match in duplicate_matches:
    print(match)
```

Remember first value is the melted_pop value, aka the correct/reference one; in other words, it should be present. Out of these duplicate matches, 
- ('SABANG', 'SUBANG'): WRONG! should be SABANG, Kota Sabang
- ('LANGSA', 'LANGKAT'): WRONG! should be LANGSA, Kota Langsa
- ('LANGKAT', 'LANGKAT') -- correct
- ('DKI JAKARTA', 'KOTA SURAKARTA'): WRONG! should be DKI JAKARTA, Daerah Khusus Ibukota Jakarta
- ('SUBANG', 'SUBANG') -- correct
- ('KOTA SURAKARTA', 'KOTA SURAKARTA') -- correct 
- ('D I YOGYAKARTA', 'KOTA YOGYAKARTA'): WRONG! should be DI YOGYAKARTA, Daerah Istimewa Yogyakarta
- ('KOTA YOGYAKARTA', 'KOTA YOGYAKARTA') -- correct

```python
# Update the matches_melted_to_area dictionary with corrected mappings
matches_melted_to_area.update({
    'SABANG': 'KOTA SABANG',
    'LANGSA': 'KOTA LANGSA',
    'DKI JAKARTA': 'DAERAH KHUSUS IBUKOTA JAKARTA',
    'D I YOGYAKARTA': 'DAERAH ISTIMEWA YOGYAKARTA'
})

# Check if there are any remaining duplicate matches
duplicate_matches = []
for key, value in matches_melted_to_area.items():
    if list(matches_melted_to_area.values()).count(value) > 1:
        duplicate_matches.append((key, value))

# Print the updated matches
print("Updated Matches:")
for match in matches_melted_to_area.items():
    print(match)
```

```python
matches_set = set(matches_melted_to_area.values())
if len(matches_set) != len(matches_melted_to_area):
    print("Duplicate matches found!")
else:
    print("No duplicate matches.")
```

```python
len(matches_melted_to_area)
```

#### Issue 2) These towns are present in melted_pop but they couldn't find a matching entry in the area dataset.

```python
# Get the unique town names from melted_pop
unique_towns_melted_pop = melted_pop['Province/Regency/City'].unique()

# Find unmatched towns
unmatched_towns = [town for town in unique_towns_melted_pop if town not in matches_melted_to_area]

# Print unmatched towns
print("Unmatched towns:")
for town in unmatched_towns:
    print(town)
```

```python
# Define the mappings
new_mappings = {
    "TOBA SAMOSIR / TOBA": "TOBA",
    "KEP SERIBU": "ADMINISTRASI KEPULAUAN SERIBU",
    "MAMUJU UTARA / PASANGKAYU": "PASANGKAYU"
}

# Add new mappings
matches_melted_to_area.update(new_mappings)

# Delete specified entries
entries_to_delete = ["INDONESIA"]
for entry in entries_to_delete:
    if entry in matches_melted_to_area:
        del matches_melted_to_area[entry]

print("New mappings added and specified entries deleted successfully.")
```

```python
len(matches_melted_to_area)
```

### Step 2) Now merge! After fixing all issues so have 548 unique matches, we expect 548 places x 6 years (2017-2022) = 3288 rows no NANs

```python
# Swap keys and values in the dictionary
matches_melted_to_area_swapped = {v: k for k, v in matches_melted_to_area.items()}

# Use the swapped dictionary for mapping
area['Matched_town'] = area['Kabupaten/kota'].map(matches_melted_to_area_swapped)

# Merge datasets based on the matches from melted_pop to area
merged_data = pd.merge(melted_pop, area, how='left', left_on='Province/Regency/City', right_on='Matched_town')

# Assuming 'Indonesia' is in the column 'Matched_town', you can drop it like this
merged_data = merged_data.drop(merged_data[merged_data['Province/Regency/City'] == 'INDONESIA'].index)

# Drop redundant columns
merged_data.drop(columns=['Kabupaten/kota','Matched_town'], inplace=True)

# Display the new merged dataset
merged_data
```

## There are NANs in plastic %! To impute, find mean or median of each province at that given year

```python
plastic_comp[plastic_comp.isna().any(axis=1)]
```

```python
# Group data by 'Year' and 'Province' and calculate the mean using agg()
mean_by_province_year = plastic_comp.groupby(['Year', 'Province'])['Plastic (%)'].agg('mean')

def impute_nan(row):
    year = row['Year']
    province = row['Province']
    if pd.isna(row['Plastic (%)']):  # Check if NaN value
        # Get mean value for the province and year
        mean_value = mean_by_province_year.loc[(year, province)]
        return mean_value
    else:
        return row['Plastic (%)']  # Return original value if not NaN

# Apply the impute_nan function to fill NaN values in 'Plastic (%)' column
plastic_comp['Plastic (%)'] = plastic_comp.apply(impute_nan, axis=1)
plastic_comp
```

```python
# Merge the datasets on the relevant columns (Year, Province, and Kabupaten/Kota)
plastic_data = pd.merge(trash_amount, plastic_comp, on=['Year', 'Province', 'Regency/City'])

# Convert 'Per day (ton)' column to numeric
plastic_data['Per day (ton)'] = plastic_data['Per day (ton)'].str.replace(',', '').astype(float)

# Multiply 'Per day (ton)' by 'Plastic (%)' to get 'Amount of plastic per day'
plastic_data['Plastic per day (ton)'] = plastic_data['Per day (ton)'] * (plastic_data['Plastic (%)'] / 100)

# Multiply 'Per year (ton)' by 'Plastic (%)' to get 'Amount of plastic per year'
plastic_data['Per year (ton)'] = plastic_data['Per year (ton)'].str.replace(',', '').astype(float)
plastic_data['Plastic per year (ton)'] = plastic_data['Per year (ton)'] * (plastic_data['Plastic (%)'] / 100)

# Select only the relevant columns
plastic_data = plastic_data[['Year', 'Province', 'Regency/City', 'Plastic per day (ton)', 'Plastic per year (ton)']]

# Display the final dataset
plastic_data
```
